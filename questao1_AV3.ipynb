```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # necess√°rio para 3D
import numpy as np

L = 5
R = 0.3

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gera todas as combina√ß√µes de coordenadas (i, j, k)
coords = [(i, j, k) for i in range(-L, L+1)
                      for j in range(-L, L+1)
                      for k in range(-L, L+1)]

for (x, y, z) in coords:
    # Desenha uma esfera no ponto (x,y,z)
    u, v = np.mgrid[0:2*np.pi:12j, 0:np.pi:6j]
    xs = x + R * np.cos(u) * np.sin(v)
    ys = y + R * np.sin(u) * np.sin(v)
    zs = z + R * np.cos(v)
    ax.plot_surface(xs, ys, zs, color='b', shade=True)

ax.set_box_aspect([1,1,1])  # para que os eixos tenham escala igual
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def fcc_lattice(length_of_side):
    L = length_of_side
    R = 0.3
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Malha da esfera (fixa para todas as esferas)
    u, v = np.mgrid[0:2*np.pi:12j, 0:np.pi:6j]
    xs_unit = R * np.cos(u) * np.sin(v)
    ys_unit = R * np.sin(u) * np.sin(v)
    zs_unit = R * np.cos(v)
    
    for i in range(-L, L+1):
        for j in range(-L, L+1):
            for k in range(-L, L+1):
                # Posi√ß√µes dos centros das esferas do FCC
                positions = [
                    (i, j, k),
                    (i+0.5, j+0.5, k),
                    (i, j+0.5, k+0.5),
                    (i+0.5, j, k+0.5)
                ]
                for (x, y, z) in positions:
                    xs = xs_unit + x
                    ys = ys_unit + y
                    zs = zs_unit + z
                    ax.plot_surface(xs, ys, zs, color='purple', shade=True, linewidth=0, antialiased=False)

    ax.set_box_aspect([1,1,1])  # para manter propor√ß√£o igual nos eixos
    plt.show()

fcc_lattice(2)
```
